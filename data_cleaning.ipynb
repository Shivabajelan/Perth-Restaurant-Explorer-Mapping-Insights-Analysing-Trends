```python
# Dependencies
import pandas as pd
```

```python
# Read the data into a Pandas DataFrame
yelp_data = pd.read_csv('data/yelp_data.csv')
yelp_data.head()
```

```python
# Get a brief summary of the yelp_data DataFrame.
yelp_data.info()
```

```python
# Create a new column 'id' with integer IDs starting from 0
yelp_data['id'] = range(len(yelp_data))

# Set the 'id' column as the index
yelp_data.set_index('id', inplace=True)

# Filter rows where the state is 'WA' 
yelp_data = yelp_data[yelp_data['location.state'] == 'WA']
```

```python
# Creating the desired address from the available columns
# Define a function takes a row of a DataFrame as input and concatenates the address parts in the specified order.
def concatenate_address(row):
    address_parts = [
        row['location.address1'],
        row['location.address2'],
        row['location.address3'],
        row['location.city'],
        row['location.zip_code']
    ]
    # Filter out empty parts in the columns and join them with comma separator
    # Using a list comprehension to iterate over each part of the address stored in the address_parts list. 
    full_address = ', '.join(part for part in address_parts if pd.notnull(part))
    return full_address

# Apply the custom function to each row
yelp_data['full_address'] = yelp_data.apply(concatenate_address, axis=1)

# Display the DataFrame with the new 'full_address' column
yelp_data.head()
```

```python
# Drop unwanted columns
columns_to_drop = ['alias', 'phone', 'is_closed','transactions', 'location.address1', 'location.address2', 'location.address3', 'location.state', 'location.country', 'location.display_address' ]
yelp_data_1 = yelp_data.drop(columns_to_drop, axis = 1)
yelp_data_1.head()
```

```python
# Drop rows with NaN values in any column
yelp_data_1.dropna(inplace=True)

# Round values in the 'distance' column to two decimal places
yelp_data_1['distance'] = yelp_data_1['distance'].round(2)
```

```python
# Extracting titles and categories from categories column and store them in separated columns

# Define extract_alias function
def extract_alias(categories):
    for category in categories:
        if 'alias' in category:
            return category['alias']
    return None

# Apply the function to the 'categories' column
# We use the eval() function to convert the string representations of dictionaries in the 'categories' column into actual dictionaries. 
yelp_data_1['alias'] = yelp_data_1['categories'].apply(lambda x: extract_alias(eval(x)))

# efine extract_titles function
def extract_titles(categories):
    titles = []
    for category in categories:
        if 'title' in category:
            titles.append(category['title'])
    return titles if titles else None

# Apply the function to the 'categories' column
yelp_data_1['titles'] = yelp_data_1['categories'].apply(lambda x: extract_titles(eval(x)))


# Print out the DataFrame head to verify new columns creation
yelp_data_1.head()
```

```python
# Drop categories and alias columns
yelp_data_1 = yelp_data_1.drop(['categories', 'alias'], axis = 1)
yelp_data_1.head()
```

```python
# Get unique values in the 'location.city' column
unique_cities = yelp_data_1['location.city'].unique()

# Print unique cities
print(unique_cities)
```

```python
# Rename the columns
yelp_data_1 = yelp_data_1.rename(columns = {'name':'business_name', 'display_phone':'phone', 'coordinates.latitude':'latitude', 'coordinates.longitude':'longitude','location.city':'location', 'location.zip_code' :'postcode', 'full_address':'address', 'titles':'categories'})
yelp_data_1
```

```python
# Rearrange columns in the DataFrame
yelp_data_1_clean = yelp_data_1[['business_name', 'categories', 'rating', 'review_count', 'price', 'image_url', 'url', 'phone', 'distance', 'latitude', 'longitude', 'location',
       'postcode', 'address' ]]
yelp_data_1_clean.head()
```

```python
# Get a brief summary of the yefinal clean DataFrame.
yelp_data_1_clean.info()
```

```python
# Export the clean DataFrame as a CSV file
yelp_data_1_clean.to_csv("data/yelp_data_clean.csv")
```
