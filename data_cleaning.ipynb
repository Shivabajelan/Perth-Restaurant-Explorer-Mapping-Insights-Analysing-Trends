```python
# Dependencies
import pandas as pd
import json
```

```python
# Read the data into a Pandas DataFrame
yelp_data = pd.read_csv('data/yelp_data.csv')
yelp_data.head()
```

```python
# Get a brief summary of the yelp_data DataFrame.
yelp_data.info()
```

```python
# Extract ethnicits from categories column
import json
# replace the single quotes ('') with "" so they can be read with json.loads
categories = [i.replace("'",'"') for i in yelp_data.categories]
# get all title types
all_cats = []
for row in categories:
    # load data as json format
    data = json.loads(row)
    # get list of unique categories
    for dictionary in data:
        if dictionary['title'] not in all_cats:
            all_cats.append(dictionary['title'])
# get list of relavent cuisines (manually selected the ones I wanted from all_cats list)
relavent = ['Latin American',
 'Italian',
 'Tapas/Small Plates',
 'Szechuan',
 'Chinese',
 'Australian',
 'Modern Australian',
 'Korean',
 'Scandinavian',
 'Shanghainese',
 'Dim Sum',
 'Cantonese',
 'American',
 'Spanish',
 'Japanese',
 'Sushi Bars',
 'Izakaya',
 'Lebanese',
 'Arabic',
 'Mexican',
 'Vietnamese',
 'Asian Fusion',
 'Singaporean',
 'Thai',
 'Malaysian',
 'Indonesian',
 'Indian',
 'Cambodian',
 'Middle Eastern',
 'Greek',
 'Mediterranean',
 'French',
 'Moroccan',
 'Taiwanese',
 'Peruvian',
 'Brazilian',
 'Modern European',
 'Burmese',
 'Irish',
 'African',
 'Afghan',
 'Persian/Iranian',
 'Himalayan/Nepalese',
 'Argentine']
# loop through all the rows, convert to json and append only the relevant categories
all_titles = []
for row in categories:
    data = json.loads(row)
    row_titles = []
    for dictionary in data:
        # if the title is in the relevant cuisine type, add it to the list otherwise leave out
        if dictionary['title'] in relavent:
            row_titles.append(dictionary['title'])
    # concat all the list items into a string with '+' as a separator
    all_titles.append(' + '.join(row_titles))
# update the categories column with the new strings
yelp_data['categories'] = all_titles
yelp_data.head()
```

```python
# Check the number of empty cells in the categories column
num_empty_cells = yelp_data['categories'].isnull().sum()
print("Number of empty cells in the categories column:", num_empty_cells)
```

```python
# Create a new column 'id' with integer IDs starting from 0
yelp_data['id'] = range(len(yelp_data))

# Set the 'id' column as the index
yelp_data.set_index('id', inplace=True)

# Filter rows where the state is 'WA' 
yelp_data = yelp_data[yelp_data['location.state'] == 'WA']
```

```python
# Creating the desired address from the available columns
# Define a function takes a row of a DataFrame as input and concatenates the address parts in the specified order.
def concatenate_address(row):
    address_parts = [
        row['location.address1'],
        row['location.address2'],
        row['location.address3'],
        row['location.city'],
        row['location.zip_code']
    ]
    # Filter out empty parts in the columns and join them with comma separator
    # Using a list comprehension to iterate over each part of the address stored in the address_parts list. 
    full_address = ', '.join(part for part in address_parts if pd.notnull(part))
    return full_address

# Apply the custom function to each row
yelp_data['full_address'] = yelp_data.apply(concatenate_address, axis=1)

# Display the DataFrame with the new 'full_address' column
yelp_data.head()
```

```python
# Change the price level from $ to an integer 
price_mapping = {'$': 1, '$$': 2, '$$$': 3, '$$$$': 4}
yelp_data['price_level'] = yelp_data['price'].map(price_mapping)
```

```python
# Drop unwanted columns
columns_to_drop = ['alias', 'phone', 'is_closed','transactions', 'location.address1', 'location.address2', 'location.address3', 'location.state', 'location.country', 'location.display_address' ]
yelp_data_1 = yelp_data.drop(columns_to_drop, axis = 1)
yelp_data_1.head()
```

```python
# Drop rows with NaN values in any column
yelp_data_1.dropna(inplace=True)
yelp_data_1.head()
```

```python
# Round values in the 'distance' column to two decimal places
yelp_data_1['distance'] = yelp_data_1['distance'].round(2)
```

```python
# Rename the columns
yelp_data_1 = yelp_data_1.rename(columns = {'name':'business_name', 'display_phone':'phone', 'coordinates.latitude':'latitude', 'coordinates.longitude':'longitude','location.city':'location', 'location.zip_code' :'postcode', 'full_address':'address'})
yelp_data_1.head()
```

```python
# Rearrange columns in the DataFrame
yelp_data_1_clean = yelp_data_1[['business_name', 'categories', 'rating', 'review_count', 'price', 'price_level', 'image_url', 'url', 'phone', 'distance', 'latitude', 'longitude', 'location',
       'postcode', 'address' ]]
yelp_data_1_clean.head()
```

```python
# Get a brief summary of the final clean DataFrame.
yelp_data_1_clean.info()
```

```python
# Export the clean DataFrame as a CSV file
yelp_data_1_clean.to_csv("data/yelp_data_clean.csv")
```

```python
# Export the clean DataFrame as a Excel file
yelp_data_1_clean.to_excel("data/yelp_data_clean.xlsx")
```
