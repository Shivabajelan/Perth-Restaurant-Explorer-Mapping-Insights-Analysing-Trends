```python
# Dependencies
import pandas as pd
```

```python
# Read the data into a Pandas DataFrame
yelp_data = pd.read_csv('data/yelp_data.csv')
yelp_data.head()
```

```python
# Get a brief summary of the yelp_data DataFrame.
yelp_data.info()
```

```python
# Create a new column 'id' with integer IDs starting from 0
yelp_data['id'] = range(len(yelp_data))

# Set the 'id' column as the index
yelp_data.set_index('id', inplace=True)
```

```python
# Drop unwanted columns
columns_to_drop = ['alias', 'phone', 'is_closed','transactions', 'location.address2', 'location.address3', 'location.zip_code', 'location.state', 'location.country' ]
yelp_data_1 = yelp_data.drop(columns_to_drop, axis = 1)
yelp_data_1.head()
```

```python
# Drop rows with NaN values in any column
yelp_data_1.dropna(inplace=True)
```
